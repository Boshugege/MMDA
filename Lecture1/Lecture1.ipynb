```python
# coding lab 1

import numpy as np
import scipy.linalg as la

def kctb_matrices(n):
    a = np.zeros(n)
    a[0], a[1] = 2, -1
    K = la.toeplitz(a)
    
    C = K.copy()
    C[0, -1] = -1
    C[-1, 0] = -1
    C = C

    T = K.copy()
    T[0, 0] = 1
    T = T

    B = K.copy()
    B[0, 0] = 1
    B[-1, -1] = 1
    B = B

    return K, C, T, B

n = 4
K, C, T, B = kctb_matrices(n)
print("K=\n", K)
print("C=\n", C)
print("T=\n", T)
print("B=\n", B)
```

```python
# coding lab 2

def solve_hanging_system(n):
    K, C, T, B = kctb_matrices(n)
    
    H = K.copy()
    H[n-1,n-1] = 1

    f = np.ones(n)
    return la.solve(H, f)

solve = solve_hanging_system(100)
print(solve)
    
```
