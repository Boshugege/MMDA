### Problem 1 (Gaussian Elimination)

由于 $A=LU$，原方程可写为

$ LUx=f $

令 $ Ux=y $

则 $ Ly=f $

设 $ y=\begin{bmatrix}
y_1\\y_2\\y_3
\end{bmatrix} $

则

$\begin{bmatrix}
1&0&0\\
-1&1&0\\
0&-1&1
\end{bmatrix}\begin{bmatrix}
y_1\\y_2\\y_3
\end{bmatrix}=\begin{bmatrix}
1\\1\\1
\end{bmatrix}.$

得到方程组

$
\begin{cases}
y_1=1\\
-y_1+y_2=1\\
-y_2+y_3=1
\end{cases}
$

因此

$
y=
\begin{bmatrix}
1\\2\\3
\end{bmatrix}.
$

设

$
x=
\begin{bmatrix}
x_1\\x_2\\x_3
\end{bmatrix}
$

则

$\begin{bmatrix}
1&-1&0\\
0&1&-1\\
0&0&1
\end{bmatrix}\begin{bmatrix}
x_1\\x_2\\x_3
\end{bmatrix}=\begin{bmatrix}
1\\2\\3
\end{bmatrix}.$

得到

$
\begin{cases}
x_1-x_2=1\\
x_2-x_3=2\\
x_3=3
\end{cases}
$

回代：

$
x_3=3
$

$
x_2=2+3=5
$

$
x_1=1+5=6
$

故答案为：

$
x=
\begin{bmatrix}
6\\
5\\
3
\end{bmatrix}.
$

#### Optional

对于一般 $n$ 阶情形：

$
y_k=y_{k-1}+1,\quad y_1=1
$

因此

$
y_k=k.
$

$
x_k-x_{k+1}=k
$

得到

$
x_k=\sum_{j=k}^n j.
$

即

$
x_k=\frac{(n-k+1)(n+k)}{2}.
$

因此

$
x=
\begin{bmatrix}
1+2+\cdots+n\\
2+3+\cdots+n\\
\vdots\\
n
\end{bmatrix}.
$

### Problem 2 (Reflection Matrix)

#### (a)

先验证 $Hu=-u$。

由
$
u=
\begin{bmatrix}
\cos\theta\\
0\\
\sin\theta
\end{bmatrix},
$
可得
$
Hu=
\begin{bmatrix}
-\cos 2\theta & 0 & -\sin 2\theta\\
0 & 1 & 0\\
-\sin 2\theta & 0 & \cos 2\theta
\end{bmatrix}
\begin{bmatrix}
\cos\theta\\
0\\
\sin\theta
\end{bmatrix}.
$

第一分量为
$
-\cos 2\theta\cos\theta-\sin 2\theta\sin\theta
= -\cos(2\theta-\theta)
= -\cos\theta.
$

第二分量为
$
0.
$

第三分量为
$
-\sin 2\theta\cos\theta+\cos 2\theta\sin\theta
= \sin(\theta-2\theta)
= -\sin\theta.
$

因此
$
Hu=
\begin{bmatrix}
-\cos\theta\\
0\\
-\sin\theta
\end{bmatrix}
=-u.
$

下面证明
$
H=I-2uu^\top.
$

先计算
$
uu^\top=
\begin{bmatrix}
\cos\theta\\
0\\
\sin\theta
\end{bmatrix}
\begin{bmatrix}
\cos\theta & 0 & \sin\theta
\end{bmatrix} = \begin{bmatrix}
\cos^2\theta & 0 & \cos\theta\sin\theta\\
0 & 0 & 0\\
\sin\theta\cos\theta & 0 & \sin^2\theta
\end{bmatrix}.
$

于是
$
I-2uu^\top = \begin{bmatrix}
1-2\cos^2\theta & 0 & -2\sin\theta\cos\theta\\
0 & 1 & 0\\
-2\sin\theta\cos\theta & 0 & 1-2\sin^2\theta
\end{bmatrix}.
$

利用恒等式
$
1-2\cos^2\theta=-\cos 2\theta,\qquad
1-2\sin^2\theta=\cos 2\theta,\qquad
2\sin\theta\cos\theta=\sin 2\theta,
$

得到
$
I-2uu^\top=
\begin{bmatrix}
-\cos 2\theta & 0 & -\sin 2\theta\\
0 & 1 & 0\\
-\sin 2\theta & 0 & \cos 2\theta
\end{bmatrix}
=H.
$

故
$
Hu=-u,\qquad H=I-2uu^\top.
$

#### (b)

设两块镜面的法向量分别为
$
u_\alpha=
\begin{bmatrix}
\cos\alpha\\
0\\
\sin\alpha
\end{bmatrix},
\qquad
u_\beta=
\begin{bmatrix}
\cos\beta\\
0\\
\sin\beta
\end{bmatrix}.
$

则相应反射矩阵为
$
H_\alpha=
\begin{bmatrix}
-\cos 2\alpha & 0 & -\sin 2\alpha\\
0 & 1 & 0\\
-\sin 2\alpha & 0 & \cos 2\alpha
\end{bmatrix},
\qquad
H_\beta=
\begin{bmatrix}
-\cos 2\beta & 0 & -\sin 2\beta\\
0 & 1 & 0\\
-\sin 2\beta & 0 & \cos 2\beta
\end{bmatrix}.
$

计算复合变换 $H_\alpha H_\beta$：

$
H_\alpha H_\beta=
\begin{bmatrix}
\cos(2\alpha-2\beta) & 0 & \sin(2\alpha-2\beta)\\
0 & 1 & 0\\
-\sin(2\alpha-2\beta) & 0 & \cos(2\alpha-2\beta)
\end{bmatrix}.
$

这正是绕 $y$-轴的旋转矩阵，记作
$
R_\gamma=
\begin{bmatrix}
\cos\gamma & 0 & \sin\gamma\\
0 & 1 & 0\\
-\sin\gamma & 0 & \cos\gamma
\end{bmatrix},
$
其中
$
\gamma=2(\alpha-\beta).
$

因此
$
H_\alpha\circ H_\beta=R_\gamma,\qquad \gamma=2(\alpha-\beta).
$

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.io import wavfile
from sklearn.decomposition import FastICA

def save_wav(filename: str, signal: np.ndarray, sample_rate: int = 16000) -> None:
    signal_int16 = np.clip(signal * 32767, -32768, 32767).astype(np.int16)
    wavfile.write(filename, sample_rate, signal_int16)


def plot_signal(signal: np.ndarray, title: str, filename: str) -> None:
    plt.figure(figsize=(10, 4))
    plt.plot(signal, linewidth=1.0)
    plt.title(title)
    plt.xlabel("Sample index")
    plt.ylabel("Amplitude")
    plt.tight_layout()
    plt.savefig(filename, dpi=200)
    plt.close()


def choose_representative_waveform(rank1_matrix: np.ndarray) -> np.ndarray:
    row_norms = np.linalg.norm(rank1_matrix, axis=1)
    idx = int(np.argmax(row_norms))
    return rank1_matrix[idx, :]


def corr_abs(a: np.ndarray, b: np.ndarray) -> float:
    a = np.asarray(a).ravel()
    b = np.asarray(b).ravel()
    a = a - np.mean(a)
    b = b - np.mean(b)
    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)
    if na == 0 or nb == 0:
        return 0.0
    return abs(np.dot(a, b) / (na * nb))

def main():
    csv_path = Path("summed_100.csv")
    X = np.loadtxt(csv_path, delimiter=",")

    print(f"Loaded mixed matrix X with shape: {X.shape}")

    sample_rate = 16000

    # Part (a)(b): SVD
    U, S, VT = np.linalg.svd(X, full_matrices=False)

    print("\n=== SVD ===")
    print("First 10 singular values:")
    print(S[:10])

    X1 = S[0] * np.outer(U[:, 0], VT[0, :])
    X2 = S[1] * np.outer(U[:, 1], VT[1, :])

    svd_source_1 = choose_representative_waveform(X1)
    svd_source_2 = choose_representative_waveform(X2)

    plot_signal(svd_source_1, "SVD Reconstructed Source 1", "svd_source_1.png")
    plot_signal(svd_source_2, "SVD Reconstructed Source 2", "svd_source_2.png")

    save_wav("svd_source_1.wav", svd_source_1, sample_rate=sample_rate)
    save_wav("svd_source_2.wav", svd_source_2, sample_rate=sample_rate)

    X_rank2 = X1 + X2
    rel_err_svd_rank2 = np.linalg.norm(X - X_rank2, ord="fro") / np.linalg.norm(X, ord="fro")
    print(f"Relative Frobenius error of rank-2 SVD approximation: {rel_err_svd_rank2:.6f}")

    # Optional Part (c): ICA
    print("\n=== ICA ===")
    ica = FastICA(n_components=2, random_state=0, max_iter=5000, tol=1e-4)
    ica.fit(X)
    A_ica = ica.mixing_

    print(f"ICA mixing matrix shape: {A_ica.shape}")

    # For X shaped (samples, time_features), columns of A_ica are temporal source waveforms.
    ica_source_1 = A_ica[:, 0]
    ica_source_2 = A_ica[:, 1]

    plot_signal(ica_source_1, "ICA Separated Source 1", "ica_source_1.png")
    plot_signal(ica_source_2, "ICA Separated Source 2", "ica_source_2.png")

    save_wav("ica_source_1.wav", ica_source_1, sample_rate=sample_rate)
    save_wav("ica_source_2.wav", ica_source_2, sample_rate=sample_rate)

    # Optional Part (d): Compare ICA and SVD
    c11 = corr_abs(svd_source_1, ica_source_1)
    c12 = corr_abs(svd_source_1, ica_source_2)
    c21 = corr_abs(svd_source_2, ica_source_1)
    c22 = corr_abs(svd_source_2, ica_source_2)

    print("\nAbsolute correlations between SVD and ICA outputs:")
    print(f"corr(|svd1, ica1|) = {c11:.6f}")
    print(f"corr(|svd1, ica2|) = {c12:.6f}")
    print(f"corr(|svd2, ica1|) = {c21:.6f}")
    print(f"corr(|svd2, ica2|) = {c22:.6f}")

    print("ICA的分离程度显著高于SVD，可能是因为源未必正交？")

if __name__ == "__main__":
    main()

    from IPython.display import Image, display

    image_files = [
        "svd_source_1.png",
        "svd_source_2.png",
        "ica_source_1.png",
        "ica_source_2.png",
    ]

    for image_name in image_files:
        image_path = Path(image_name)
        if image_path.exists():
            print(image_name)
            display(Image(filename=str(image_path)))
```
