```python
import numpy as np
from time import perf_counter

n = 1000

def build_K(n: int) -> np.ndarray:
    main_diag = 2 * np.ones(n)
    off_diag = -1 * np.ones(n - 1)
    return np.diag(main_diag) + np.diag(off_diag, 1) + np.diag(off_diag, -1)

K = build_K(n)
f = np.zeros(n)
f[0] = 4

repeats = 5

solve_times = []
solve_result = None
for _ in range(repeats):
    start = perf_counter()
    solve_result = np.linalg.solve(K, f)
    solve_times.append(perf_counter() - start)

inv_times = []
inv_result = None
for _ in range(repeats):
    start = perf_counter()
    inv_result = np.linalg.inv(K) @ f
    inv_times.append(perf_counter() - start)

solve_time = min(solve_times)
inv_time = min(inv_times)
solution_difference = np.linalg.norm(solve_result - inv_result, ord=np.inf)
residual_norm = np.linalg.norm(K @ solve_result - f, ord=np.inf)

print(f"n = {n}")
print(f"np.linalg.solve runtime (best of {repeats}): {solve_time:.6f} s")
print(f"np.linalg.inv(K) @ f runtime (best of {repeats}): {inv_time:.6f} s")
print(f"speed ratio (inv / solve): {inv_time / solve_time:.2f}")
print(f"max-norm difference between solutions: {solution_difference:.3e}")
print(f"max-norm residual of solve result: {residual_norm:.3e}")
print("first 10 entries of u from solve:")
print(solve_result[:10])
```
